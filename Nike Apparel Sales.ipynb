```python
import pandas as pd
import matplotlib.pyplot as mt
import seaborn as sea
```

```python
nk=pd.read_csv(r'C:\Users\hp\OneDrive\Desktop\Nike_sales.csv')
nk
```

```python
nk.isnull().sum().sum()
```

```python
nk.dropna(inplace=True)
```

```python
nk.shape
```

```python
nk.head(3)
```

```python
nk.drop(['Retailer ID','Invoice Date'],axis=1,inplace=True)
```

```python
nk
```

```python
nk.dtypes
```

```python
nk[['Retailer','Region','State','City','Product']]=nk[['Retailer','Region','State','City','Product']].astype('string')
nk['Price per Unit']=nk['Price per Unit'].astype(int)
```

```python
nk.dtypes
```

```python
nk.head(3)
```

```python
nk['Quantity of Unit Sold']=nk['Units Sold']/nk['Price per Unit']
```

```python
nk
```

```python
nk.duplicated().sum()
```

```python
nk.drop_duplicates(keep='first',inplace=True)
```

```python
nk.duplicated().sum()
```

```python
nk['Quantity of Unit Sold']=nk['Quantity of Unit Sold'].astype(int)
```

```python
nk.head(2)
```

```python
nk['Region'].value_counts()
```

```python
def region(reg):
    if reg=='Midwest':
        return 'mid-west'.title()
    elif reg=='Northeast':
        return 'north-east'.title()
    elif reg=='Southeast':
        return 'south-east'.title()
    else:return reg
```

```python
nk['Region']=nk['Region'].apply(region)
```

```python
mt.style.use('fivethirtyeight')
sea.countplot(data=nk,x='Product',palette='Set2')
mt.xticks(rotation=90,fontweight='heavy',size=10)
mt.yticks(fontweight='heavy',size=10)
```

```python
sea.set_style('darkgrid')
sea.set_context('notebook')
sea.color_palette('pastel')
# ec=sea.color_palette('magma')
mt.figure(figsize=(12,6))
sales_product=sea.barplot(data=nk,y='Product',
                          x='Total Sales',errorbar=('ci', 0),
                          width=0.5,palette='magma')

mt.title('Average Sales by each Product',size=20,fontweight='semibold')

mt.xticks(fontweight='heavy',size=15)
mt.yticks(fontweight='demibold',size=12)
mt.xlabel('Average Sales',fontweight='medium',size=15)
mt.ylabel('Products',fontweight='medium',size=15)

for sp in sales_product.containers:
    sales_product.bar_label(sp,fontweight='bold',size=18,padding=-60,color='white')
```

```python
sales_region=nk.groupby('Region')['Total Sales'].mean().sort_values(ascending=False)
sales_region
```

```python
sales_retailer=nk.groupby('Retailer')['Total Sales'].mean().sort_values(ascending=False)
sales_retailer
```

```python
mt.style.available
```

```python
mt.figure(figsize=(15,10))
mt.style.use('ggplot')
palette=sea.color_palette('OrRd')
mt.subplot(2,2,1)
sales_region.plot(kind='pie',startangle=-7,
                  autopct='%0.2f%%',shadow=True,
                  colors=palette, 
                  wedgeprops={'edgecolor':'black',
                              'linewidth':0.5},
                  textprops={'fontweight':'semibold',
                             'size':15},radius=1.2)

mt.title('Sales by Region',fontweight='bold',size=15)
mt.ylabel('')

mt.subplot(2,2,2)
palette_2=sea.color_palette("pastel")
sales_retailer.plot(kind='pie',counterclock=False,
                    startangle=200,autopct='%0.2f%%',
                    shadow=True,colors=palette_2,
                   wedgeprops={'edgecolor':'black','linewidth':0.5},textprops={'fontweight':'semibold','size':15},radius=1.2)

mt.title('Sales Distribution by Vendors',fontweight='bold',size=15)
mt.ylabel('')
```
